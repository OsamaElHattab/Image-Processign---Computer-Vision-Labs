```python
from commonfunctions import *
import math
import numpy as np
import skimage.io as io
from skimage.filters import median
from skimage.filters import gaussian
from skimage.util import random_noise


%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
'''
(1) Your own median filter algorithm
'''
img = io.imread('pout.tif')
io.imshow(img)
io.show()

noisy_img = random_noise(img, mode='s&p', amount=0.05)
io.imshow(noisy_img)
io.show()

window_width = 3
window_height = 3
edgex = window_width // 2
edgey = window_height // 2

output_img = np.zeros_like(img)

for x in range(edgex, img.shape[1] - edgex):
    for y in range(edgey, img.shape[0] - edgey):
        colorArray = []
        for fx in range(window_width):
            for fy in range(window_height):
                colorArray.append(img[y + fy - edgey, x + fx - edgex])
        colorArray.sort()
        output_img[y, x] = colorArray[len(colorArray) // 2]


print("My Own Median Filter")
io.imshow(output_img)
io.show()
```

```python
'''
(2) skimage median filter
'''

ski_median_img = median(image=noisy_img)
print("Skimage Median Filter")
io.imshow(ski_median_img)
io.show()

```

```python
'''
(3) Gaussion Filters
'''

ski_guaussian_img_8 = gaussian(noisy_img, sigma=8)
print("Skimage Gaussian Filter sigmq = 8")
io.imshow(ski_guaussian_img_8)
io.show()

ski_guaussian_img_3 = gaussian(noisy_img, sigma=3)
print("Skimage Gaussian Filter sigma = 3")
io.imshow(ski_guaussian_img_3)
io.show()

ski_guaussian_img_02 = gaussian(noisy_img, sigma=0.2)
print("Skimage Gaussian Filter sigma = 0.2")
io.imshow(ski_guaussian_img_02)
io.show()

```
